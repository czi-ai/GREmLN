# Load and formatting

```python
raw_count.obs['cell_type'] = np.array(cell_info['CellType'])
raw_count.obs['sample'] = np.array(cell_info['Sample_ID'])
raw_count
```

# QC

```python
def qc_metrics(adata):
  adata.var['mt'] = adata.var_names.str.startswith('MT-')
  sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, inplace=True)
```

```python
def apply_qc_filters(adata, mito_thres=20, umi_thres=1000, umi_upper=1e6):
  before = adata.shape[0]
  adata = adata[adata.obs['pct_counts_mt'] < mito_thres, :]
  sc.pp.filter_cells(adata, min_counts=umi_thres)
  sc.pp.filter_cells(adata, max_counts=umi_upper)
  after = adata.shape[0]
  return adata, before, after
```

```python
qc_metrics(adata=raw_count)
```

```python
sc.pl.violin(raw_count, keys='total_counts', groupby='sample')
```

```python
sc.pl.violin(raw_count, keys='pct_counts_mt', groupby='sample')
```

```python
sc.pl.scatter(raw_count, x='total_counts', y='pct_counts_mt')
sc.pl.scatter(raw_count, x='total_counts', y='n_genes_by_counts')
```

```python
raw_count, before, after = apply_qc_filters(raw_count, mito_thres=20, umi_thres=1000, umi_upper=80000)
```

# Normalisation, Dim reduction, Visualisation

```python
raw_count.layers['raw_count'] = raw_count.X
raw_count
```

```python
sc.pp.normalize_total(raw_count, target_sum=1e6)
sc.pp.log1p(raw_count)
sc.pp.pca(raw_count, svd_solver='arpack')
sc.pp.neighbors(raw_count)
sc.tl.umap(raw_count)
```

```python
sc.pl.umap(raw_count, color=['cell_type', 'sample'])
```
